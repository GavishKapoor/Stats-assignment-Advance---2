Theory Questions

1. What is hypothesis testing in statistics?

Hypothesis testing is a statistical method used to make decisions about a population based on
sample data. It involves making an assumption (hypothesis), collecting data, and then determining
whether the data supports the assumption or suggests an alternative.

2. What is the null hypothesis, and how does it differ from the alternative hypothesis?

Null hypothesis (H0): A default statement that there is no effect or difference.
Alternative hypothesis (H1): A statement that contradicts the null, suggesting a real effect or
difference.

3. What is the significance level in hypothesis testing, and why is it important?

The significance level (alpha) is the probability of rejecting the null hypothesis when it is actually
true. Common values are 0.05 or 0.01. It defines the threshold for statistical significance and helps
control Type 1 errors.

4. What does a P-value represent in hypothesis testing?

The P-value is the probability of obtaining test results at least as extreme as the observed results,
assuming the null hypothesis is true.

5. How do you interpret the P-value in hypothesis testing?

If P-value alpha: Reject the null hypothesis.
If P-value > alpha: Fail to reject the null hypothesis.

6. What are Type 1 and Type 2 errors in hypothesis testing?

Type 1 error: Rejecting a true null hypothesis.
Type 2 error: Failing to reject a false null hypothesis.

7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?

One-tailed test: Tests for a deviation in one direction.
Two-tailed test: Tests for deviations in both directions from the mean.

8. What is the Z-test, and when is it used in hypothesis testing?

A Z-test is used to determine if there is a significant difference between sample and population
means when the population standard deviation is known.

9. How do you calculate the Z-score, and what does it represent in hypothesis testing?

Z = (X - mu) / sigma
Where X is the sample mean, mu is the population mean, and sigma is the standard deviation. It
measures how many standard deviations a data point is from the mean.

10. What is the T-distribution, and when should it be used instead of the normal distribution?

The T-distribution is used when the sample size is small (< 30) and the population standard
deviation is unknown

11. What is the difference between a Z-test and a T-test?

Z-test: Population standard deviation known.
T-test: Population standard deviation unknown and uses sample data.

12. What is the T-test, and how is it used in hypothesis testing?

A T-test evaluates whether the means of two groups are statistically different. It's often used for
small sample sizes.

13. What is the relationship between Z-test and T-test in hypothesis testing?

Both tests compare means, but the T-test is used when the sample size is small or variance is
unknown, while the Z-test is used for larger samples with known variance.

14. What is a confidence interval, and how is it used to interpret statistical results?

A confidence interval gives a range of values within which the true population parameter is expected
to lie, with a certain level of confidence (e.g., 95%).

15. What is the margin of error, and how does it affect the confidence interval?

The margin of error is the range added/subtracted from the sample statistic to create the confidence
interval. A larger margin implies less precision.

16. How is Bayes' Theorem used in statistics, and what is its significance?

Bayes' Theorem updates the probability of a hypothesis based on new evidence. Its crucial for
decision-making in uncertain conditions.

17. What is the Chi-square distribution, and when is it used?

The Chi-square distribution is used for testing relationships between categorical variables and for
variance analysis.

18. What is the Chi-square goodness of fit test, and how is it applied?

This test checks if a sample distribution matches an expected distribution. It compares observed vs.
expected frequencies.

19. What is the F-distribution, and when is it used in hypothesis testing?

The F-distribution is used to compare two variances or in ANOVA tests to analyze variance among
group means.

20. What is an ANOVA test, and what are its assumptions?

ANOVA (Analysis of Variance) compares means among three or more groups. Assumptions:
- Independence
- Normality
- Equal variances

21. What are the different types of ANOVA tests?

One-way ANOVA: One independent variable.
Two-way ANOVA: Two independent variables.
Repeated measures ANOVA: Same subjects across conditions.

22. What is the F-test, and how does it relate to hypothesis testing?

The F-test compares two variances and forms the basis of ANOVA. It's used to test if group
variances are significantly different.

```python
# 28
import numpy as np
from scipy.stats import skew, kurtosis

def analyze_distribution(data):
    skewness = skew(data)
    kurt = kurtosis(data)  # excess kurtosis (normal = 0)

    print("Skewness:", skewness)
    print("Kurtosis:", kurt)

# Generate normally distributed data
data = np.random.normal(loc=0, scale=1, size=1000)

# Call the function
analyze_distribution(data)
```

```python
# 4
import matplotlib.pyplot as plt
from scipy.stats import norm

z_stat, p_val = ztest(sample, value=100)
x = np.linspace(-4, 4, 300)
plt.plot(x, norm.pdf(x))
plt.axvline(x=z_stat, color='red', linestyle='--', label='Z-stat')
plt.axvline(x=norm.ppf(0.975), color='green', linestyle='--', label='Critical value')
plt.fill_between(x, norm.pdf(x), where=(x > norm.ppf(0.975)) | (x < norm.ppf(0.025)), color='gray', alpha=0.3)
plt.title("Two-tailed Z-Test")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 5
# Conceptual — simulate and plot two normal curves
x = np.linspace(-5, 10, 500)
null = norm.pdf(x, loc=0, scale=1)
alt = norm.pdf(x, loc=2, scale=1)

plt.plot(x, null, label='H0')
plt.plot(x, alt, label='H1', linestyle='--')
plt.fill_between(x, null, where=(x > 1.64), color='red', alpha=0.3, label='Type I Error')
plt.fill_between(x, alt, where=(x <= 1.64), color='blue', alpha=0.3, label='Type II Error')
plt.legend()
plt.title("Type I and Type II Errors")
plt.grid(True)
plt.show()
```

```python
# 9
import scipy.stats as stats

def confidence_interval(data, confidence=0.95):
    mean = np.mean(data)
    se = stats.sem(data)
    margin = se * stats.t.ppf((1 + confidence) / 2, len(data) - 1)
    return mean - margin, mean + margin

sample = np.random.normal(100, 10, 30)
print("Confidence Interval:", confidence_interval(sample))
```
